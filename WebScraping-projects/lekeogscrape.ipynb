# Leker med scraping av "enkle" nettsider, og lage enkel ordsky

```python
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

#### Henter inn titler fra romantiske b√∏ker 

```python
url = "https://books.toscrape.com/catalogue/category/books/science_22/index.html"
```

```python
# Fetching the webpage
response = requests.get(url)
```

```python
# Parsing the webpage with BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Finding all the book titles
book_titles = [title.text for title in soup.find_all("h3")]
```

```python
# Creating a string from the book titles
titles_string = " ".join(book_titles)
```

```python
# Generating the word cloud
wordcloud = WordCloud().generate(titles_string)
```

```python
# Plotting the word cloud
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
